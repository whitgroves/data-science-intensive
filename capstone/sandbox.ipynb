```python
import pandas as pd
```

```python
path = 'data/2016_cand.csv'
```

```python
%%time
reader = pd.read_csv(path, encoding='latin-1', chunksize=2**11)

csv_chunks = []
for csv_chunk in reader:
    csv_chunks.append(csv_chunk)
    
df = pd.concat(csv_chunks)
```

```python
df.columns
```

```python
df.head()
```

```python
import numpy as np
df = df.replace('\\N', np.NaN)
df2 = df.dropna(how='all')
```

```python
df2.groupby('RecipientCandidateNameNormalized')['TransactionAmount'].sum().sort_values(ascending=False).head(20)
```

```python
df2[df2.RecipientCandidateNameNormalized 
    == 'Hillary Clinton'][['RecipientCandidateNameNormalized',
                           'DonorNameNormalized',
                           'TransactionAmount',
                           'TransactionDate']].sort_values(by='TransactionAmount', ascending=False).head(20)
```

```python
df2[df2.RecipientCandidateNameNormalized 
    == 'Hillary Clinton'].groupby('DonorNameNormalized')['TransactionAmount'].sum().sort_values(ascending=False)
```

```python
# Which groups give the most often?
df2.groupby('DonorNameNormalized')['TransactionAmount'].sum().sort_values(ascending=False)
```

```python
df2[df2.RecipientCandidateNameNormalized 
    == 'Donald Trump'][['RecipientCandidateNameNormalized',
                            'DonorNameNormalized',
                            'TransactionAmount',
                            'TransactionDate']].sort_values(by='TransactionAmount', ascending=False).head(20)
```
